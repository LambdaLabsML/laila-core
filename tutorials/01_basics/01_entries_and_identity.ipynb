# Tutorial 1: Entries and Identity

This tutorial introduces the most fundamental concept in LAILA: the **Entry**.

You will learn how to:
- Create constant and variable entries
- Inspect `global_id`, `uuid`, `scopes`, `evolution`, `state`, and `data`
- Use nicknames for deterministic identity
- Wrap different data types

```python
import laila
import numpy as np
```

## Creating a constant entry

A constant is an immutable entry — once created, its data and identity are fixed.

```python
entry = laila.constant(data={"message": "hello from laila"})
print(entry)
```

## The global ID

Every entry has a `global_id` — a fully-qualified string that uniquely identifies it.
The format is `LAILA:<scope>:GLOBAL_ID:<uuid>`.

```python
print("global_id:", entry.global_id)
```

## Inspecting entry attributes

An entry carries several identity and state attributes:
- `uuid` — the underlying UUID
- `scopes` — list of scope strings (defaults to `['ENTRY']`)
- `evolution` — version counter (`None` for constants, `0` for variables)
- `state` — lifecycle state (`STAGED` or `READY`)
- `data` — the actual payload

```python
print("uuid:     ", entry.uuid)
print("scopes:   ", entry.scopes)
print("evolution:", entry.evolution)
print("state:    ", entry.state)
print("data:     ", entry.data)
```

## Constants vs. variables

A **constant** has no evolution counter. A **variable** starts at evolution `0` and can be versioned.

```python
c = laila.constant(data=42)
print("constant evolution:", c.evolution)
print("constant global_id:", c.global_id)

v = laila.variable(data=42)
print("\nvariable evolution:", v.evolution)
print("variable global_id:", v.global_id)
```

## Deterministic naming with nicknames

Passing a `nickname` derives the UUID deterministically — the same nickname always produces the same `global_id`.

```python
a = laila.constant(data="first", nickname="my_entry")
b = laila.constant(data="second", nickname="my_entry")

print("a global_id:", a.global_id)
print("b global_id:", b.global_id)
print("Same identity?", a.global_id == b.global_id)
print("a.data:", a.data)
print("b.data:", b.data)
```

## Wrapping different data types

`laila.constant` accepts any serializable Python object — scalars, dicts, lists, numpy arrays, and torch tensors.

```python
scalar_entry = laila.constant(data=3.14)
dict_entry = laila.constant(data={"key": "value"})
array_entry = laila.constant(data=np.zeros((5, 5)))

print("scalar:", scalar_entry.data)
print("dict:  ", dict_entry.data)
print("array shape:", array_entry.data.shape)
```

```python
import torch

tensor_entry = laila.constant(data=torch.randn(3, 224, 224))
print("tensor shape:", tensor_entry.data.shape)
print("tensor dtype:", tensor_entry.data.dtype)
```

## Summary

- `laila.constant(data=...)` creates an immutable entry
- `laila.variable(data=...)` creates a versioned entry (evolution starts at 0)
- Every entry has a `global_id` — the key for storage and retrieval
- Passing `nickname=` derives a deterministic UUID
- The `data` property gives you back the original payload
